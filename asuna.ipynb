```python
def bit_len_max(bit_len):
    return 2 ** int(bit_len) - 1

def get_trait_value(code, trait_index):
    if (bits[trait_index][1] == 0): return -1
    tv = (int(code,16) >> bits[trait_index][0]) & bit_len_max(bits[trait_index][1])
    if tv == bit_len_max(bits[trait_index][1]): return -1
    return tv
```

```python
# finds conflicts given a set of traits
def find_conflicts(traits):
    potential_conflicts_by_slot = {k:b for k,v in traits.items() if (b := banned.get(k,{}).get(v,None)) != None}
    potential_conflicts_list = [i for sl in [banned.get(k,{}).get(v,[]) for k,v in traits.items()] for i in sl]
    conflicts = [i for i, c in collections.Counter(potential_conflicts_list).items() if c > 1]
    conflicts_by_slot = {k:v for k,v in potential_conflicts_by_slot.items() if any(item in conflicts for item in v)}
    return conflicts_by_slot
    
# generate random traits and apply overrides
# if there are conflicts, try again up to 10 times
# then return it regardless because it's probably the overrides causing the conflicts
def generate_random_traits(overrides = {}):
    overrides = {k:v for k,v in overrides.items() if "random" not in v}
    max_attempts = 10
    i = 0
    conflicts = {}
    traits = {}
    while i < max_attempts:
        # roll traits
        traits = {k:random.choice([i for i in v.keys() if i != "-1"]) for k,v in asset_labels.items()}

        # some traits doesn't need to always exist, so we roll to remove some
        traits = {k:v for k,v in traits.items() if (random.random() < random_trait_exist_chance.get(k,1))}

        # apply the overrides provided by the user
        traits.update(overrides)
        
        # check conflicts
        conflicts = find_conflicts(traits)
        has_conflicts = len(conflicts.items())
        if not has_conflicts:
            return traits, {}
        i += 1

    # too many retries, return anyways with conflict
    logging.warn("""Failed to generate random traits after {} attempts. Conflicts: {}""".format(max_attempts, conflicts))
    return traits, conflicts

def override_random_traits(input_traits):
    max_attempts = 10
    i = 0
    conflicts = {}
    random_slots = [k for k,v in input_traits.items() if v in ["random", "random_none"]]
    random_none_slots = [k for k,v in input_traits.items() if v=="random_none"]
    traits = input_traits.copy()
    while i < max_attempts:
        # roll traits
        random_traits = {k:random.choice([i for i in v.keys() if i != "-1"]) for k,v in asset_labels.items() if k in random_slots}
        random_traits = {k:v for k,v in random_traits.items() if (k not in random_none_slots) or (random.random() < random_trait_exist_chance.get(k,1))}

        # apply the overrides provided by the user
        traits.update(random_traits)
        traits = {k:(v if (v != "random_none") else "-1") for k,v in traits.items()}
        
        # check conflicts
        conflicts = find_conflicts(traits)
        has_conflicts = len(conflicts.items())
        if not has_conflicts:
            return {k:v for k,v in traits.items() if k in random_slots}
        i += 1
    
    # too many retries, return anyways with conflict
    logging.warn("""Failed to generate override random traits after {} attempts. Conflicts: {}""".format(max_attempts, conflicts))
    return random_traits

# given asuna id or random, plus a set of overrides, return a set of traits
def get_asuna(overrides, token_id=None):
    if token_id == None:
        return generate_random_traits(overrides)
    elif isinstance(token_id,str):
        traits = code_to_traits(token_id)
        traits.update({"5":"0", "14":"0"})
    else:
        r = requests.get(f'NO_PERMISSION_TO_RESHARE_API_URL')
        traits = {k:str(v) for k,v in r.json()[str(token_id)]['traitData'].items()}
    random_traits = override_random_traits(overrides)
    overrides.update(random_traits)
    traits.update(overrides)
    conflicts = find_conflicts(traits)
    return traits, conflicts

# given a set of traits, return the text version of traits
def get_readable_traits(t, c, o):
    return {traits_labels[int(k)]:asset_labels[k][v]['name'] + ("" if k not in c.keys() else " :warning:") + ("" if k not in o.keys() else " :star:") for k,v in t.items()}


def generate_image(traits):
    traits_labeled = {i:None for i in traits_labels}
    traits_labeled.update({traits_labels[int(k)]:(v,asset_labels[k][v]['name']) for k,v in traits.items()})
    img = Image.new("RGBA", (IMG_WIDTH, IMG_HEIGHT), (255, 255, 255, 0))
    for i, l in enumerate(layer_keys):
        key = [str(i)] + [traits_labeled[j][0] for j in l if traits_labeled[j] is not None]
        key = "layer/" + "/".join([k.zfill(6) for k in key])
        layers = data.get(key, [])
        for layer in layers:
            fpath = f"./images/{key.replace('/','_')}_{layer}"
            layer_img = Image.open(fpath).convert("RGBA")
            if "Multiply" in layer:
                img = Image.fromarray(np.uint8(multiply(np.array(img).astype(float), np.array(layer_img).astype(float), 1)))
            elif "Screen" in layer:
                img = Image.fromarray(np.uint8(screen(np.array(img).astype(float), np.array(layer_img).astype(float), 1)))
            else:
                img = Image.alpha_composite(img, layer_img)
                # img.paste(layer_img, (0,0), mask=layer_img)
            layer_img.close()
    # img = img.resize( (int(0.33 * s) for s in img.size) )
    return img
```
